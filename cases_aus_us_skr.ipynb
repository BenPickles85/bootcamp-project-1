```python
file_to_load = "Resources/Aus-covid-data.csv"
```

```python
read_file = pd.read_csv(file_to_load)
read_file.info()
```

```python
clean_us_df = read_file[(read_file["location"]=='United States')].reset_index( drop = True)
clean_us_df.describe()
```

```python
clean_aus_df = read_file[(read_file["location"]=='Australia')].reset_index( drop = True)
clean_aus_df.head()
```

```python
clean_skr_df = read_file[(read_file["location"]=='South Korea')].reset_index( drop = True)
clean_skr_df.head(50) 
```

```python
skr_data = clean_skr_df[["date","total_cases","new_cases"]]
skr_data.head(5)
```

```python
us_data = clean_us_df[["date","total_cases","new_cases"]]
us_data
```

```python
au_data = clean_aus_df[["date","total_cases","new_cases"]]
au_data
```

```python
merged_df = au_data.merge(us_data, how = 'inner', on='date', suffixes = ('_aus','_us'))
merged_df
```

```python
merge_three_df = merged_df.merge(skr_data, how = 'inner', on='date')
merge_three_df
```

```python
merge_three_df.to_csv("Resources/cases_au_us_skr_data.csv")
```
